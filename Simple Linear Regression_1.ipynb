```python
import pandas as pd
df = pd.read_csv(r"C:\Users\yashk\Downloads\startup_data.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
#Remove duplicates
#df.drop_duplicates()
```

```python
# Seperate X and Y features
#X: Independent feature => RND
#Y:Dependent feature=> PROFIT
```

```python
X=df[["RND"]]
Y=df[["PROFIT"]]
```

```python
#There are no missing values nor any duplicated rows in this data
#still we need to add cleaning step(for future data)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
```

```python
pipe = make_pipeline(SimpleImputer(strategy="median")).set_output(transform="pandas")
pipe
```

```python
X_pre = pipe.fit_transform(X)
X_pre.head()
```

```python
# Model Building:Linear Regression
# Equation of line: y = B0+B1x
# B0=> y intercept
# B1 => slope of the line
# ProfitPredicted = B0+B1*RND
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
#B0- y intercept
model.intercept_
```

```python
#B1- slope of the line
model.coef_
```

```python
# ProfitPredicted = 49032.9+0.854*RND
# RND=0 ProftPredicted = 49032.9
# RND =1 it will increase by 0.854times
```

```python
Y.head()
```

```python
Ypreds = model.predict(X)
Ypreds[0:5]
```

```python
# Evaluation Metrics
# Mean Squared Error
# Mean Absolute Error
# RMSE
# R2-score
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
#mean Squared Error
MSE = mean_squared_error(Y,Ypreds)
#RMSE-Root Mean squared Error
RMSE = MSE**(1/2)
#MAE
MAE = mean_absolute_error(Y,Ypreds)
#R2 score
r2 = r2_score(Y,Ypreds)
print(f"Mean Squared Error:{MSE:.2f}")
print(f"Root Mean Squared Error:{RMSE:.2f}")
print(f"Mean Absolute Error:{MAE:.2f}")
print(f"R2 score:{r2:.2f}")
```

```python
# R2 square value is greater than 0.80, we can consider this model for final model building
```

```python

# Train test Split
# rate at which you divide data:
# 100%
# 60% training 40% testing
# 70% training and 30% testing
# 80% training and 20% testing
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Train Test split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
X_train
```

```python
#Train model
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
#Prediction
y_pred = model.predict(X_test)
```

```python
#mean Squared Error
MSE = mean_squared_error(Y_test,y_pred)
#RMSE-Root Mean squared Error
RMSE = MSE**(1/2)
#MAE
MAE = mean_absolute_error(Y_test,y_pred)
#R2 score
r2 = r2_score(Y_test,y_pred)
print(f"Mean Squared Error:{MSE:.2f}")
print(f"Root Mean Squared Error:{RMSE:.2f}")
print(f"Mean Absolute Error:{MAE:.2f}")
print(f"R2 score:{r2:.2f}")
```
